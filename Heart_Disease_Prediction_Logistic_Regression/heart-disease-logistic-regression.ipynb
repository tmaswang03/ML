```python
df = pd.read_csv('/kaggle/input/heart-disease-prediction-using-logistic-regression/framingham.csv')
imp = SimpleImputer(missing_values = np.nan, strategy = 'mean') #could also use mean, median, most freq
df = pd.DataFrame(imp.fit_transform(df), columns = df.columns)
#using sklearn simple imputer to take care of null data values by replacing with mean
df
```

```python
train_input, test_input, train_output, test_output = train_test_split(df.drop('TenYearCHD', axis = 1), df['TenYearCHD'])
logistic_regression = LogisticRegression(solver = 'liblinear', max_iter = 150) 
#using liblearn as it's better suited for larger data sets
#also, lbfgs, sag and saga does not converge (newton takes too long)
logistic_regression.fit(train_input, train_output)
logistic_regression.score(test_input, test_output)
```
